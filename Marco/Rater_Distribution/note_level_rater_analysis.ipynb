# Join Ratings + Rater Factors + Status

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import glob

rating_files = sorted(glob.glob(f"{DATA_DIR}/filtered/ratings-*.tsv"))

all_stage_data = []
```

```python
for file in tqdm(rating_files):
    df = pd.read_csv(
        file,
        sep="\t",
        usecols=["noteId", "raterParticipantId",
                 "createdAtMillis", "ratingSourceBucketed"]
    )

    # merge status
    df = df.merge(
        status[["noteId",
                "timestampMillisOfFirstNonNMRStatus"]],
        on="noteId",
        how="inner"
    )

    # tag stage
    df["stage"] = np.where(
        df["createdAtMillis"] <
        df["timestampMillisOfFirstNonNMRStatus"],
        "pre",
        "post"
    )

    # merge rater factor
    df = df.merge(
        raterFactor[["raterParticipantId",
                     "internalRaterFactor1"]],
        on="raterParticipantId",
        how="inner"
    )

    all_stage_data.append(df)

ratings_all = pd.concat(all_stage_data, ignore_index=True)
```

```python
print(ratings_all.shape)

print(ratings_all.isna().sum())
```

`timestampMillisOfFirstNonNMRStatus` contains many missing values. This indicates that a large share of notes have not yet transitioned out of NMR status, likely because the observation window is recent or the notes have not received enough ratings to reach a consensus.

```python

ratings_all = ratings_all.dropna(
    subset=["timestampMillisOfFirstNonNMRStatus", "internalRaterFactor1"]
).copy()

print(ratings_all.shape)
```

# Bucket on 3 quantiles of internalRaterFactor1

```python
row_order = ["Interested + Nudged Raters", "Nudged Raters", "Interested Raters"]
col_order = ["pre", "post"]
col_titles = {"pre": "Before status", "post": "After status"}


# Create factor buckets [low, mid, high] using 3 quantiles
ratings_all["factor_bucket"] = pd.qcut(
    ratings_all["internalRaterFactor1"],
    q=3,
    labels=["Low", "Mid", "High"],
    duplicates="drop"
)

bucket_order = ["Low", "Mid", "High"]

ratings_all["factor_bucket"] = pd.Categorical(
    ratings_all["factor_bucket"],
    categories=bucket_order,
    ordered=True
)
```

## Questions

### 1) Why do *Nudged* users appear in the **Post** stage?

**Stage definition**
- `pre`: `rating.createdAtMillis < note.timestampMillisOfFirstNonNMRStatus`
- `post`: `rating.createdAtMillis >= note.timestampMillisOfFirstNonNMRStatus`

**Key point**
- `stage` is determined by the **time of the rating** relative to the note’s first non-NMR timestamp.
- So if a **POPULATION_SAMPLED (Nudged)** rating is submitted **after** the note crosses the first non-NMR threshold, it will be labeled **post**.

**In short**
- Nudged users can exist in **post** because nudged ratings can be submitted **after** the status-change cutoff.



# Note Level Entropy

```python
import numpy as np

def normalized_entropy(values, bins=10):
    if len(values) < 5:
        return np.nan

    hist, _ = np.histogram(values, bins=bins)
    p = hist / hist.sum()
    p = p[p > 0]

    H = -np.sum(p * np.log(p))
    H_norm = H / np.log(bins)
    return H_norm
```

Bin = 10

```python
group_cols = ["noteId", "stage", "ratingSourceBucketed", "factor_bucket"]

# Single Note Level (note × stage × rater type × bucket),
# compute the normalized entropy of internalRaterFactor1

# This measures how diverse the rater factor distribution is within that group
entropy_df = (
    ratings_all
    .groupby(group_cols)["internalRaterFactor1"]
    .apply(lambda x: normalized_entropy(x, bins=10))
    .reset_index(name="entropy")
)

# Average entropy across notes within each
# (bucket × stage × rater type) combination

# mean within-note rater diversity by condition
avg_entropy = (
    entropy_df
    .groupby(["factor_bucket", "stage", "ratingSourceBucketed"], observed=True)["entropy"]
    .mean()
    .reset_index()
)

print(avg_entropy)
```

```python
plot_entropy = avg_entropy.copy()

plot_entropy["source_label"] = plot_entropy["ratingSourceBucketed"].map({
    "DEFAULT": "Interested",
    "POPULATION_SAMPLED": "Nudged"
})
```

```python
g = sns.catplot(
    data=plot_entropy,
    kind="bar",
    x="stage",
    y="entropy",
    hue="source_label",
    col="factor_bucket",
    col_order=["Low", "Mid", "High"],
    order=["pre", "post"],
    height=4,
    aspect=0.9,
    sharey=True,
    legend_out=True
)

g.set_axis_labels("Stage", "Average Normalized Entropy")
g.set_titles("Bucket: {col_name}")

g.fig.suptitle(
    "Average Within-Note Rater Factor Entropy",
    fontsize=14
)

g.fig.subplots_adjust(top=0.85)

for ax in g.axes.flat:
    ax.set_ylim(0, 1)
    for p in ax.patches:
        h = p.get_height()
        if h > 0:
            ax.annotate(
                f"{h:.3f}",
                (p.get_x() + p.get_width()/2., h),
                ha="center", va="bottom",
                fontsize=9,
                xytext=(0, 3),
                textcoords="offset points"
            )

g._legend.set_bbox_to_anchor((1.05, 1))
g._legend.set_loc("upper left")

plt.show()
```

# Note Level Variance

```python
group_cols = ["noteId", "stage", "ratingSourceBucketed", "factor_bucket"]


# For each (note × stage × rater type × bucket),
# compute standard deviation and variance of internalRaterFactor1.

# This measures how spread out raters are within a note under each condition.
var_df = (
    ratings_all
    .groupby(group_cols)["internalRaterFactor1"]
    .agg(["std", "var"])
    .reset_index()
)


# Average dispersion across notes within each
# (factor_bucket × stage × rater type) condition.

# This moves from note-level spread to condition-level average spread.
avg_var = (
    var_df
    .groupby(["factor_bucket", "stage", "ratingSourceBucketed"], observed=True)
    .mean()
    .reset_index()
)
```

```python
plot_var = avg_var.copy()

plot_var["source_label"] = plot_var["ratingSourceBucketed"].map({
    "DEFAULT": "Interested",
    "POPULATION_SAMPLED": "Nudged"
})
```

```python
g = sns.catplot(
    data=plot_var,
    kind="bar",
    x="stage",
    y="std",
    hue="source_label",
    col="factor_bucket",
    col_order=["Low", "Mid", "High"],
    order=["pre", "post"],
    height=4,
    aspect=0.9,
    sharey=True,
    legend_out=True
)

g.set_axis_labels("Stage", "Average Standard Deviation")
g.set_titles("Bucket: {col_name}")

for ax in g.axes.flat:
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(
                f"{height:.3f}",
                (p.get_x() + p.get_width()/2, height),
                ha="center",
                va="bottom",
                fontsize=9
            )

g._legend.set_loc("upper left")
g._legend.set_bbox_to_anchor((1.05, 1))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

palette = sns.color_palette()
color_no = palette[0]
color_flip = palette[1]

bucket_order = ["Low", "Mid", "High"]

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

for i, bucket in enumerate(bucket_order):

    ax = axes[i]
    df_bucket = note_entropy[note_entropy["factor_bucket"] == bucket]

    # KDE - No Flip
    sns.kdeplot(
        data=df_bucket[df_bucket["flip"] == 0],
        x="entropy",
        fill=True,
        alpha=0.4,
        color=color_no,
        ax=ax,
        label="No Flip" if i == 0 else None
    )

    # KDE - Flip
    sns.kdeplot(
        data=df_bucket[df_bucket["flip"] == 1],
        x="entropy",
        fill=True,
        alpha=0.4,
        color=color_flip,
        ax=ax,
        label="Flip" if i == 0 else None
    )

    # Medians
    median_no = df_bucket[df_bucket["flip"] == 0]["entropy"].median()
    median_flip = df_bucket[df_bucket["flip"] == 1]["entropy"].median()

    ax.axvline(median_no, linestyle="--", linewidth=2, color=color_no)
    ax.axvline(median_flip, linestyle="--", linewidth=2, color=color_flip)

    ax.set_title(f"{bucket} Bucket")
    ax.set_xlabel("Pre-stage Normalized Entropy")
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

axes[0].set_ylabel("Density")

# Global title
fig.suptitle(
    "Pre-stage Entropy Distribution by Flip Status Across Factor Buckets",
    fontsize=14
)

# Put legend once
axes[0].legend()

plt.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show()
```

```python
import numpy as np
from scipy.stats import mannwhitneyu

for bucket in ["Low", "Mid", "High"]:

    df_bucket = note_entropy[note_entropy["factor_bucket"] == bucket]

    group0 = df_bucket[df_bucket["flip"] == 0]["entropy"]
    group1 = df_bucket[df_bucket["flip"] == 1]["entropy"]

    group0 = group0.replace([np.inf, -np.inf], np.nan).dropna()
    group1 = group1.replace([np.inf, -np.inf], np.nan).dropna()

    print(f"\nBucket: {bucket}")
    print("No Flip n =", len(group0))
    print("Flip n =", len(group1))

    if len(group0) > 0 and len(group1) > 0:
        u, p = mannwhitneyu(group0, group1, alternative="two-sided")

        print("No Flip median =", group0.median())
        print("Flip median =", group1.median())
        print("p-value =", p)
    else:
        print("Insufficient data")
```

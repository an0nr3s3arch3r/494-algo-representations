```python
from google.colab import drive

drive.mount("/content/drive/")

BASE_DIR = "/content/drive/MyDrive"
FOLDERNAME = "INFO494"
DATA_DIR = f"{BASE_DIR}/{FOLDERNAME}"
```

```python
import pandas as pd
```

```python
note1 = pd.read_csv(f"{DATA_DIR}/notes-00000.tsv", sep="\t")

note2 = pd.read_csv(f"{DATA_DIR}/notes-00001.tsv", sep="\t")
```

```python
# Create a filtered directory if it doesn't exist
import os
filtered_dir = f"{DATA_DIR}/filtered"
os.makedirs(filtered_dir, exist_ok=True)
```

```python
# Filter collaborative notes only
colab_note1 = note1[note1["isCollaborativeNote"] == 1]
colab_note2 = note2[note2["isCollaborativeNote"] == 1]

colab_note = pd.concat([colab_note1, colab_note2], ignore_index=True)

# change createdAtMillis to datetime
colab_note["createdAtMillis"] = pd.to_datetime(colab_note["createdAtMillis"], unit="ms")

colab_note.to_csv(f"{DATA_DIR}/filtered/colab_note.csv", index=False)
```

```python
# filter ratings for collaborative notes only
colab_note_ids = set(colab_note["noteId"])
rating_dfs = []
for i in range(7):
    path = f"{DATA_DIR}/ratings-{i:05d}.tsv"
    df = pd.read_csv(path, sep="\t", low_memory=False)
    df = df[df["noteId"].isin(colab_note_ids)]
    rating_dfs.append(df)
ratings = pd.concat(rating_dfs, ignore_index=True)
ratings.to_csv(f"{DATA_DIR}/filtered/ratings_all.csv", index=False)
```

```python
status = pd.read_csv(f"{DATA_DIR}/noteStatusHistory-00000.tsv", sep="\t")
# filter noteid in colab_note_ids
status = status[status["noteId"].isin(colab_note_ids)]
# change createdAtMillis and all the column started with timestamp to datetime
status["createdAtMillis"] = pd.to_datetime(status["createdAtMillis"], unit="ms")
for col in status.columns:
    if col.startswith("timestamp"):
        status[col] = pd.to_datetime(status[col], unit="ms")
status = status.drop(columns=["createdAtMillis"])
status.to_csv(f"{DATA_DIR}/filtered/note_status.csv", index=False)
```
